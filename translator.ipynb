```python
from tkinter import *
from tkinter import ttk
from googletrans import Translator, LANGUAGES

def change(text="type", src="English", dest="Hindi"):
    text1 = text
    src1 = src
    dest1 = dest
    trans = Translator()
    trans1 = trans.translate(text1, src=src1, dest=dest1)
    return trans1.text

def data():
    s = comb_sor.get()
    d = comb_dest.get()
    msg = sor_txt.get(1.0, END)
    textget = change(text=msg, src=s, dest=d)
    dest_txt.delete(1.0, END)
    dest_txt.insert(END, textget)

root = Tk()
root.title("Devash Translator")
root.geometry("900x800")
root.config(bg="red")

# Title Label
lab_title = Label(root, text="Devash Translator", font=("Times New Roman", 40, "bold"), bg="red", fg="white")
lab_title.place(x=100, y=20, height=100, width=700)

# Frame for Source and Destination Text
frame = Frame(root, bg="white", bd=2, relief="groove")
frame.place(x=50, y=170, width=800, height=600)

# Source Text Label
lab_source = Label(frame, text="Source Text", font=("Times New Roman", 20, "bold"), fg="black", bg="white")
lab_source.place(x=10, y=10, height=30, width=780)

# Source Text Area
sor_txt = Text(frame, font=("Times New Roman", 16), wrap=WORD)
sor_txt.place(x=10, y=50, height=220, width=780)

# ComboBox for selecting source language
list_text = list(LANGUAGES.values())
comb_sor = ttk.Combobox(frame, values=list_text, font=("Times New Roman", 16))
comb_sor.place(x=20, y=280, height=30, width=220)
comb_sor.set("english")

# Translate Button
button_change = Button(frame, text="TRANSLATE", relief="raised", command=data, font=("Times New Roman", 16))
button_change.place(x=250, y=280, height=30, width=150)

# ComboBox for selecting target language
comb_dest = ttk.Combobox(frame, values=list_text, font=("Times New Roman", 16))
comb_dest.place(x=420, y=280, height=30, width=218)
comb_dest.set("hindi")

# Destination Text Label
lab_dest = Label(frame, text="Dest Text", font=("Times New Roman", 20, "bold"), fg="black", bg="white")
lab_dest.place(x=10, y=320, height=30, width=780)

# Destination Text Area
dest_txt = Text(frame, font=("Times New Roman", 24), wrap=WORD)
dest_txt.place(x=10, y=360, height=220, width=780)

# Run the main loop
root.mainloop()
```
